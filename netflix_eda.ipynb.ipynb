# Netflix Exploratory Data Analysis

## Project Objective

The objective of this project is to perform Exploratory Data Analysis (EDA) on Netflix's content dataset and uncover meaningful insights related to content distribution, release trends, ratings, and geographical availability. The analysis helps understand Netflix's content strategy and global reach using Python, Pandas, Matplotlib, and Seaborn.

## Technologies Used

- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn
- Jupyter Notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/netflix_titles.csv')
df.head()
```

## Dataset Information

- Source: Netflix Titles Dataset
- Total Records: 8,807
- Total Columns: 12
- Dataset contains information about movies and TV shows available on Netflix.

## Dataset Overview

```python
df.shape
```

```python
df.info()
```

### Observation

- The dataset contains 8,807 records and 12 columns.
- Most columns are categorical (object type).
- The release_year column is numerical.
- The dataset contains information about titles, countries, ratings, genres, and release years.

## Missing Values Analysis

```python
df.isnull().sum()
```

### Observation

- The director column has the highest number of missing values (2,634).
- The cast and country columns also contain missing values.
- Only a few missing values are present in date_added, rating, and duration.
- Most of the dataset is complete and suitable for analysis.

## Movies vs TV Shows Analysis

```python
df['type'].value_counts()
```

### Key Insights

- Movies account for approximately 70% of Netflix's catalog.
- TV Shows represent a smaller but significant portion of available content.
- Netflix appears to prioritize movie production and acquisition.

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title('Comparison of Movies and TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

## Country Analysis

```python
df['country'].value_counts().head(10)
```

### Observation

- The United States has the highest number of Netflix titles (2,818).
- India ranks second with 972 titles.
- Most Netflix content originates from a few major countries.
- The platform has a strong international presence with content from many regions.

```python
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries by Number of Netflix Titles')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

## Release Year Analysis

```python
plt.figure(figsize=(12,5))
df['release_year'].value_counts().sort_index().plot()
plt.title('Distribution of Netflix Content by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

### Observation

- Netflix content production increased significantly after 2010.
- The highest number of titles were released between 2017 and 2020.
- Recent years show rapid growth in Netflix's content library.

## Rating Analysis

```python
df['rating'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Content-Producing Countries on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

### Observation

- TV-MA is the most common rating on Netflix.
- Mature and adult-oriented content forms a significant portion of the catalog.
- Netflix offers content across multiple age groups and audience categories.

## Key Findings

- Movies significantly outnumber TV Shows on Netflix.
- The United States contributes the largest amount of content.
- India is the second-largest content contributor.
- Content production increased rapidly after 2010.
- TV-MA is the most common content rating.
- Netflix maintains a diverse international content library.

## Conclusion

This exploratory analysis of Netflix content revealed important trends in content distribution, ratings, and production. Movies form the majority of Netflix's catalog, while the United States and India emerge as the leading content contributors. The analysis also highlights Netflix's rapid expansion in content production after 2010 and its strong focus on mature-audience programming. Overall, Netflix demonstrates a broad and globally diverse content portfolio.
